## كم عدد الطلاب الذي جابوا الدرجه الكامله

```python
reading_full = df[df['reading_score'] == 100]['reading_score'].count()
writing_full = df[df['writing_score'] == 100]['writing_score'].count()
math_full = df[df['math_score'] == 100]['math_score'].count()
3
print(f'Number of students with full marks in Reading: {reading_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Math: {math_full}')
```

## كم عدد الطلاب الذي نجحوا

```python
df['math_result'] = df['math_score'].apply(lambda x: 'successful' if x >= 50 else 'Failed')
df['reading_result'] = df['reading_score'].apply(lambda x: 'successful' if x >= 50 else 'Failed')
df['writing_result'] = df['writing_score'].apply(lambda x: 'successful' if x >= 50 else 'Failed')


print(df['math_result'].value_counts())

print(df['reading_result'].value_counts())

print( df['writing_result'].count())
```

```python
print(df['nationality'].value_counts())
```

```python
plt.subplot(1, 3, 1)
sns.countplot(x=df['math_result'])
plt.title('Math Result')


plt.subplot(1, 3, 2)
sns.countplot(x=df['reading_result'])
plt.title('Reading Result')

plt.subplot(1, 3, 3)
sns.countplot(x=df['writing_result'])
plt.title('Writing Result')

plt.tight_layout()
plt.show()


pass_math_count = df['math_result'].value_counts()
print(pass_math_count)
pass_math_count = df['reading_result'].value_counts()
print(pass_math_count)
pass_math_count = df['writing_result'].value_counts()
print(pass_math_count)
```

### متوسط درجات الطلاب

```python
plt.subplot(1, 3, 1)
sns.histplot(x=df['math_score'])
plt.title('Math Score')


plt.subplot(1, 3, 2)
sns.histplot(x=df['reading_score'])
plt.title('Reading Score')

plt.subplot(1, 3, 3)
sns.histplot(x=df['writing_score'])
plt.title('Writing Score')

plt.tight_layout()
plt.show()


math_mean = df['math_score'].mean()
reading_mean = df['reading_score'].mean()
writing_mean = df['writing_score'].mean()

print(f'Average score in Math : { math_mean }\nAverage score in Reading : { reading_mean }\nAverage score in Writing : { writing_mean }\n\n')
```

```python
plt.figure(figsize=(10, 8)) 
plt.subplot(2, 3, 1)
x = df['gender'].value_counts()
labels = 'female', 'male'
plt.pie(x,  labels = labels, autopct = '%.2f%%')
plt.title('Gender')


plt.subplot(2, 3, 2)
x = df['nationality'].value_counts()
labels = 'Indian', 'Chinese', 'Yemeni', 'African', 'European'
plt.pie(x, labels = labels, autopct = '%.2f%%')
plt.title('nationality')


plt.subplot(2, 3, 3)
x = df['lunch'].value_counts()
labels = 'Standard', 'Free'
plt.pie(x, labels = labels, autopct = '%.2f%%')
plt.title('Lunch')


plt.subplot(2, 3, 4)
x = df['test_preparation_course'].value_counts()
labels = 'None', 'Completed'
plt.pie(x, labels = labels, autopct = '%.2f%%')
plt.title('Test Course')


plt.subplot(2, 3, 5)
x = df['parent_education'].value_counts()
labels = 'Some College', "Associate's Degree",'High School','Some High School',"Bachelor's Degree","Master's Degree"
plt.pie(x,labels = labels, autopct = '%.2f%%')
plt.title('Parental Education')
plt.show()
```

### أي الفئتين حصلت على أعلى درجة: الطلاب أم الطالبات؟


```python


plt.subplot(1, 3, 1)
sns.boxplot(x=df['gender'], y=df['math_score'])
plt.title('Math score vs Gender')

plt.subplot(1, 3, 2)
sns.boxplot(x=df['gender'], y=df['reading_score'])
plt.title('Reading score vs Gender')

plt.subplot(1, 3, 3)
sns.boxplot(x=df['gender'], y=df['writing_score'])
plt.title('Writing score vs Gender')

plt.tight_layout()
plt.show()

```

### هل يؤثر المستوى التعليمي للوالدين على الأداء الأكاديمي للأبناء؟


```python
plt.subplot(1, 3, 1)
sns.boxplot(x=df['parent_education'], y=df['math_score'])
plt.title('Math score vs Parents education')
plt.xticks(rotation=90)

plt.subplot(1, 3, 2)
sns.boxplot(x=df['parent_education'], y=df['reading_score'])
plt.title('Reading score vs Parents education')
plt.xticks(rotation=90)

plt.subplot(1, 3, 3)
sns.boxplot(x=df['parent_education'], y=df['writing_score'])
plt.title('Writing score vs Parents education')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
